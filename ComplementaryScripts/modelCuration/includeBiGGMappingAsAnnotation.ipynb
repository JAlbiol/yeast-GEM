```python
from cobra.io import load_yaml_model, read_sbml_model
import pandas as pd
from cobra import show_versions
```

```python
show_versions()
```

```python
yeast = load_yaml_model("../../ModelFiles/yml/yeastGEM.yml")
```

```python
yeast = read_sbml_model("../../ModelFiles/xml/yeastGEM.xml")
```

```python
global_met_dictionary = pd.read_csv("../../ComplementaryData/modelCuration/globalMetDictionary.csv", names=['default.metabolite', 'bigg.metabolite'] )
```

```python
global_met_dictionary = global_met_dictionary.set_index('default.metabolite').to_dict()['bigg.metabolite']
```

```python
global_rxn_dictionary = pd.read_csv("../../ComplementaryData/modelCuration/globalRxnDictionary.csv", names=['default.reaction', 'bigg.reaction'] )
```

```python
global_rxn_dictionary = global_rxn_dictionary.set_index('default.reaction').to_dict()['bigg.reaction']
```

```python
global_met_dictionary
```

```python
for met in yeast.metabolites:
    met_id = met.id.replace("__91__", '[').replace("__93__", ']')
    if met_id in global_met_dictionary.keys():
        met.annotation["bigg.metabolite"] = global_met_dictionary[met_id].split('[')[0]
        met.annotation["default.metabolite"] = met.id
```

```python
for rxn in yeast.reactions:
    if rxn.id in global_rxn_dictionary.keys():
        rxn.annotation["bigg.reaction"] = global_rxn_dictionary[rxn.id]
        rxn.annotation["default.reaction"] = rxn.id
```
